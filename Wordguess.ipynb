```python
import random
from hangman import draw_hangman

# Word bank
words = ['spartan', 'cedar', 'football', 'programming', 'computer', 'business', 'accounting', 'broad', 'major', 'minor', 'graded', 'project', 'algorithm', 'branching', 'nested']
def main():
    # Seed input
    while True:
        seed_choice = input("Would you like to enter a seed? (y/n): ").lower()
        if seed_choice == 'y':
            while True:
                try:
                    seed = int(input("Please enter a seed: "))
                    if seed >= 0:
                        random.seed(seed)
                        break
                    else:
                        print("ERROR: Invalid seed.")
                except ValueError:
                    print("ERROR: Invalid seed.")
            break
        elif seed_choice == 'n':
            break
        else:
            print("Error: answer must be y or n")

    # Word selection
    word = random.choice(words)
    word_length = len(word)

    # Game setup
    tiles = ['_'] * word_length
    guessed_letters = []
    incorrect_guesses = 5
    remaining_letters = word_length

    # Main game loop
    while incorrect_guesses > 0 and remaining_letters > 0:
        draw_hangman(tiles, f"Letters remaining: {remaining_letters:^30}", incorrect_guesses, ', '.join(guessed_letters))

        guess = input("Enter a letter: ").lower()

        if not guess.isalpha() or len(guess) != 1:
            print("Invalid input. Please enter a single letter.")
            continue

        if guess in guessed_letters:
            print("You have already guessed that letter!")
            continue

        guessed_letters.append(guess)

        if guess in word:
            for i in range(word_length):
                if word[i] == guess:
                    tiles[i] = guess
                    remaining_letters -= 1
        else:
            incorrect_guesses -= 1

    # Game over
    draw_hangman(tiles, f"Letters remaining: {remaining_letters:^30}", incorrect_guesses, ', '.join(guessed_letters))

    if remaining_letters == 0:
        print(f"You win!\nThe word was {word}!")
    else:
        print(f"You lost!\nThe word was {word}!")

if __name__ == "__main__":
    main()
```

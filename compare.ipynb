```python
import random
import string  # Added import for string module
from hangman import draw_hangman

# Word bank
words = ['programming', 'algorithm', 'function', 'variable', 'loop', 'conditional']

def main():
    # Seed input
    while True:
        seed_choice = input("Would you like to enter a seed? (y/n):\n").lower()
        if seed_choice == 'y':
            while True:
                try:
                    seed = int(input("Please enter a seed:\n"))
                    if seed >= 0:
                        random.seed(seed)
                        break
                    else:
                        print("ERROR: Invalid seed.")
                except ValueError:
                    print("ERROR: Invalid seed.")
            break
        elif seed_choice == 'n':
            break
        else:
            print("Error: answer must be y or n")

    # Word selection
    word = random.choice(words)
    word_length = len(word)

    # Game setup
    tiles = ['_'] * word_length
    guessed_letters = []
    incorrect_guesses = 5
    remaining_letters = word_length

    # Main game loop
    while incorrect_guesses > 0 and remaining_letters > 0:
        draw_hangman(tiles, f"Letters remaining: {remaining_letters:^15}", incorrect_guesses, ', '.join(guessed_letters))

        guess = input("Enter a letter:\n").lower()

        # Check if input is valid (Changed the condition)
        if guess not in string.ascii_lowercase or len(guess) != 1:
            print("Invalid input. Please enter a single letter.")
            continue

        if guess in guessed_letters:
            print("You have already guessed that letter!")
            continue

        guessed_letters.append(guess)

        if guess in word:
            for i in range(word_length):
                if word[i] == guess:
                    tiles[i] = guess
                    remaining_letters -= 1
        else:
            incorrect_guesses -= 1

    # Game over
    draw_hangman(tiles, f"Letters remaining: {remaining_letters:^15}", incorrect_guesses, ', '.join(guessed_letters))

    if remaining_letters == 0:
        print(f"You win!\nThe word was {word}!")
    else:
        print(f"You lost!\nThe word was {word}!")

if __name__ == "__main__":
    main()
```

```python
import random
from hangman import draw_hangman

def main():
    words = ["programming", "python", "hangman", "example", "words"]
    
    print("Would you like to enter a seed? (y/n):")
    while True:
        seed_choice = input().lower()
        if seed_choice in ['y', 'n']:
            break
        print("Error: answer must be y or n")
        print("Would you like to enter a seed? (y/n):")

    if seed_choice == 'y':
        while True:
            print("Please enter a seed:")
            seed = int(input())
            if seed >= 0:
                break
            print("ERROR: Invalid seed.")

        random.seed(seed)

    word = random.choice(words)
    guessed_letters = set()
    correct_letters = set()
    incorrect_guesses_remaining = 5
    hidden_word = ['_' for _ in word]

    while incorrect_guesses_remaining > 0:
        remaining_letters = sum([1 for letter in word if letter not in correct_letters])
        guessed_string = ', '.join(sorted(guessed_letters))
        draw_hangman(hidden_word, f"Letters remaining: {remaining_letters:^30}", incorrect_guesses_remaining, guessed_string)

        print("Enter a letter:")
        guess = input().lower()

        if guess in guessed_letters:
            print("You have already guessed that letter!")
        elif guess in word:
            guessed_letters.add(guess)
            correct_letters.add(guess)
            for i, letter in enumerate(word):
                if letter == guess:
                    hidden_word[i] = guess
        else:
            guessed_letters.add(guess)
            incorrect_guesses_remaining -= 1

        if '_' not in hidden_word:
            break

    draw_hangman(hidden_word, f"Letters remaining: {remaining_letters:^30}", incorrect_guesses_remaining, guessed_string)

    if incorrect_guesses_remaining > 0:
        print("You win!")
    else:
        print("You lost!")
    print(f"The word was {word}!")

if __name__ == "__main__":
    main()
```

```python
import random

def get_seed():
    while True:
        seed_choice = input("Would you like to enter a seed? (y/n):\n")
        if seed_choice.lower() == 'y':
            while True:
                seed = input("Please enter a seed:\n")
                try:
                    seed_value = int(seed)
                    if seed_value >= 0:
                        return seed_value
                    else:
                        print("ERROR: Invalid seed.")
                except ValueError:
                    print("ERROR: Invalid seed.")
        elif seed_choice.lower() == 'n':
            return None
        else:
            print("Error: answer must be y or n")

def display_hangman(guesses_remaining):
    hangman_stages = [
        "",
        """
|   .---.                      |
|  |     |                     |
|  |     |                     |
|  |     |                     |
|   '---'                      |""",
        """
|   .---.                      |
|  | o   |                     |
|  |     |                     |
|  |     |                     |
|   '---'                      |""",
        """
|   .---.                      |
|  | o o |                     |
|  |     |                     |
|  |     |                     |
|   '---'                      |""",
        """
|   .---.                      |
|  | o o |                     |
|  |  >  |                     |
|  |     |                     |
|   '---'                      |""",
        """
|   .---.                      |
|  | o o |                     |
|  |  >  |                     |
|  | '-' |                     |
|   '---'                      |""",
    ]

    if guesses_remaining <= 4:
        return hangman_stages[4 - guesses_remaining + 1]
    else:
        return hangman_stages[0]

def display_game_state(hidden_word, remaining_letters, guesses_remaining, guessed_letters):
    hangman = display_hangman(guesses_remaining)
    hidden_word_str = " ".join(hidden_word)
    boundary_length = max(len(hidden_word_str) + 8, 32)
    boundary_line = "+" + "-" * (boundary_length - 2) + "+"
    
    padding_left = " " * ((boundary_length - len(hidden_word_str)-2) // 2)
    padding_right = " " * ((boundary_length - len(hidden_word_str) -1) // 2)

    guessed_letters_str = ",".join(guessed_letters)
    guessed_padding_left = " " * ((boundary_length - len(guessed_letters_str)-2) // 2)
    guessed_padding_right = " " * ((boundary_length - len(guessed_letters_str) - 1) // 2)

    print()
    print(boundary_line)
    if guesses_remaining < 5:
        print(hangman[1:])
    else:
        print("|" + " " * (boundary_length - 2) + "|")
    print("|" + padding_left + hidden_word_str + padding_right + "|")
    print("|" + " " * (boundary_length - 2) + "|")
    print("|" + " " * 5 + f"Letters remaining: {remaining_letters}" + " " * 5 + "|")
    print("|" + " " * (boundary_length - 2) + "|")
    print("|" + " " * 5 + f"Guesses remaining: {guesses_remaining}" + " " * 5 + "|")
    print("|" + " " * (boundary_length - 2) + "|")
    print("|" + " " * 7 + "Guessed letters" + " " * 7 + " |")
    print("|" + " " * 7 + "---------------" + " " * 7 + " |")
    print("|" + guessed_padding_left + guessed_letters_str + guessed_padding_right + "|")
    print(boundary_line)
    print()




def main():
    word_bank = ['spartan', 'cedar', 'football', 'programming', 'computer', 'business', 'accounting', 'broad', 'major', 'minor', 'graded', 'project', 'algorithm', 'branching', 'nested']
    seed = get_seed()
    if seed is not None:
        random.seed(seed)

    word = random.choice(word_bank)
    hidden_word = ["_" for _ in word]
    guessed_letters = []
    guesses_remaining = 5
    remaining_letters = len(word)

    while remaining_letters > 0 and guesses_remaining > 0:
        display_game_state(hidden_word, remaining_letters, guesses_remaining, guessed_letters)
        guess = input("Enter a letter:\n").lower()

        if guess in guessed_letters:
            print("You have already guessed that letter!\n")
        elif guess in word:
            guessed_letters.append(guess)
            for i, letter in enumerate(word):
                if letter == guess:
                    hidden_word[i] = letter
                    remaining_letters -= 1
        else:
            guessed_letters.append(guess)
            guesses_remaining -= 1

    if remaining_letters == 0:
        print("You win!")
    else:
        print("You lost!")

    print(f"The word was {word}!")

if __name__ == "__main__":
    main()
```
